```python
def url_split(Url):
    url = Url
    url_main = url.split("?")[0]
    url_params = url.split("?")[1]
    url_params_2 = url_params.split("&")
    params = {}
    for i in url_params_2:
        params[i.split("=")[0]] = i.split("=")[1]
    return params,url_main
```
