```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup

from fake_useragent import UserAgent
import re
```

```python
def url_split(Url):
    url = Url
    url_main = url.split("?")[0]
    url_params = url.split("?")[1]
    url_params_2 = url_params.split("&")
    params = {}
    for i in url_params_2:
        params[i.split("=")[0]] = i.split("=")[1]
    return params,url_main
```

```python
url_main = 'http://www.korea.kr/briefing/speechList.do'
title = []
date = []
url = []
presenter = []
url_sub = "http://www.korea.kr/archive/speechView.do?"
for page in range(1,22):
    params = {
        'cateId': '2',
        'pageIndex': page,
        'srchKeyword': '',
        'srchType': 'title'
    }
    html = requests.get(url_main,params=params).text
    soup = BeautifulSoup(html,"html.parser")
    for tag in soup.select(".table_list_type01 tbody tr"):
        title_2 = tag.find("td",class_="subject").find("a").text
        title.append(title_2)
        
        url_2 = tag.find("td",class_="subject").find("a")["href"]
        url_all = url_sub + url_2
        url.append(url_all)
        
        presenter_2 = tag.find_all("td")[2].text
        presenter.append(presenter_2)
        
        date_2 = tag.find_all("td")[3].text
        date.append(date_2)
        #print(date_2)
```

```python
text = [] 
for ur in url:
    html_2 = requests.get(ur).text
    soup_2 = BeautifulSoup(html_2,"html.parser")
    for tag in soup_2.select(".view_cont"):
        text_2 = tag.text
        text.append(text_2)
```

```python
print(len(text),len(title),len(date),len(url),len(presenter))
```

```python
data = {"title":title,"date":date,"url":url,"presenter":presenter,"text":text}
dataframe = pd.DataFrame(data,columns={"title":title,"date":date,"url":url,"presenter":presenter,"text":text})
dataframe
```
