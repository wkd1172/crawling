```python
def daum_pageindex(max_page,sub):
    
    main_url = "http://bbs1.agora.media.daum.net/gaia/do/debate/list"


    index_id = {"자유": "D003" ,
               "정치" : "D101",
               "경제" : "D115",
               "부동산" : "D125",
               "주식펀드" : "D126",
               "사회" : "D110",
               "교육" : "D102",
               "문화연예" : "D109"}
    Title = []
    Date = []
    Url = []
    Like = []
    Dislike = []
    Traffic = [] 

    for page in range(1,int(max_page)+1):
        params = {
            "bbsId": index_id[sub],
            "pageIndex" : page
        }

        html_1 = requests.get(main_url,params = params).text
        soup_1 = BeautifulSoup(html_1,"html.parser")
        url_1 = "http://bbs1.agora.media.daum.net/gaia/do/debate/"

        for tag in soup_1.select(".gLlist tr"):
            try:
                #title = tag.findall("td",class_ = "gLsbj").find("a").text
                title = tag.find("td",class_="gLsbj").find("a").text
                date = tag.find("td",class_="gLdate").text
                url_sub = tag.find("td",class_="gLsbj").find("a")["href"]
                url_all = url_1 + url_sub
                like_1 = tag.find("td",class_="gLup").text
                dislike_1 = tag.find("td",class_="gLdown").text
                traffic_1 = tag.find("td",class_="gLcount").text
                Title.append(title)
                Date.append(date)
                Url.append(url_all)
                Like.append(like_1)
                Dislike.append(dislike_1)
                Traffic.append(traffic_1)
                #print(title)
                #print(tag.text)
                #print(traffic_1)
            except AttributeError:
                pass
    data = {"Title" : Title,"Date":Date,"Url":Url,"Like":Like,"Dislike":Dislike,"Traffic":Traffic}
    dataframe = pd.DataFrame(data)
    return dataframe
            
            
```

```python
daum_pageindex(10,"문화연예")
```

## 2010.09.01 페이지 찾기

#### 보니까 일정페이지 index 넘어가면 url 형식이 바뀐다. 
    
    1) 일단 바뀌는 페이지 찾기 ( 바뀌는 마지막 -> title 이 반복될때 멈추기 )
        -> 그떄 다음 누르는 페이지 -> url 찾기
    2) 그때 부터 다음페이지 의 url 추출 하는 것 하기!


```python
main_url = "http://bbs1.agora.media.daum.net/gaia/do/debate/list"
page_1 = 1
title_b = "b"
title_a = "a"
while title_b != title_a:

    params = {
        "pageIndex" : page_1,
        "bbsId" : "D102"
    }
    response = requests.get(main_url,params=params)
    html = response.text
    soup = BeautifulSoup(html,"html.parser")

    for tag in soup.select(".gLlist tr"):
        try:
            title_b = tag.find("td",class_="gLsbj").find("a").text
        
        except AttributeError:
            pass          
        

    page_2 = page_1 + 1
    params = {
        "pageIndex" : page_2,
        "bbsId" : "D102"
    }
    response = requests.get(main_url,params=params)
    html = response.text
    soup = BeautifulSoup(html,"html.parser")
    for tag in soup.select(".gLlist tr"):
        try:
            title_a = tag.find("td",class_="gLsbj").find("a").text
        
        except AttributeError:
            pass   
    page_1 += 1
    print(page_1)
print(page_1)
```
