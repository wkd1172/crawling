```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
def url_split(Url):
    url = Url
    url_main = url.split("?")[0]
    url_params = url.split("?")[1]
    url_params_2 = url_params.split("&")
    params = {}
    for i in url_params_2:
        params[i.split("=")[0]] = i.split("=")[1]
    return params,url_main
```

```python
params,url_main=url_split("http://news.donga.com/Series/70040100000001?p=1&prod=news&ymd=&m=")
```

```python
url_main = 'http://news.donga.com/Series/70040100000001'
title = []
date = []
for num in range(1,430):
    params = {'m': '',
              'p': int(20(num-1)+1),
              'prod': 'news',
              'ymd': ''}
   
    html = requests.get(url_main,params=params).text
    soup = BeautifulSoup(html,"html.parser")

    for tag in soup.select(".rightList a"):
        title_2 = tag.find("span",class_= "tit").text
        date_2 = tag.find("span",class_="date").text
        title.append(title_2)
        date.append(date_2)
data = {"title" : title,"date" : date}
dataframe = pd.DataFrame(data)
dataframe
```
