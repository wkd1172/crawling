```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup

from fake_useragent import UserAgent
import re
```

```python
def url_split(Url):
    url = Url
    url_main = url.split("?")[0]
    url_params = url.split("?")[1]
    url_params_2 = url_params.split("&")
    params = {}
    for i in url_params_2:
        params[i.split("=")[0]] = i.split("=")[1]
    return params,url_main
```

```python
params,url_main=url_split("http://srchdb1.chosun.com/pdf/i_service/pdf_SearchList.jsp?requestSearchPage=simple&collectionType=gisa&FV=%5B%BB%E7%BC%B3%5D&INDEX_FV=TI&AU_FV=&PD_F0=year&PD_TYPE=false&PD_F1=20100901&PD_OP=1&PD_F2=20181109&DATA_SORT=1&LIST_TYPE=true&sRowsperPage=50")
```

```python
p = re.compile("조선일보.\r")
```

```python
p = re.compile("조선일보.")
title = []
date = []
for page in range(1,30):
    url_1 = url.format(page)
    #print(url_1)
    html = requests.get(url_1).text
    soup = BeautifulSoup(html,"html.parser")
    try:
        for tag in soup.select("#search_list"):
            title_2 = tag.find("span",class_="list_tit").find("a").text
            title_3 = title_2.replace("\n","").replace("\r","").replace("\t","")
            title.append(title_3)
            
            date_2 = tag.find("div",class_="gigoja").text.replace("[발행일] 조선일보 ","")[0:10]
            date.append(date_2)
            
    except AttributeError:
        pass
data = {"title": title,"date" : date}
dataframe = pd.DataFrame(data)
dataframe
```
